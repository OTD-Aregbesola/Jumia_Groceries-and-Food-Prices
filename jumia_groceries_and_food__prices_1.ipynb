```python
# import the data

import pandas as pd

df = pd.read_csv('/content/store_df_multi_page.csv')
```

```python
#view the shape of the data
df.shape
```

```python
# view the 5 rows of the data
df.head()
```

```python
#convert the columns to lowercase
df = df.rename( columns= {'Name' : 'name', 'Current Price ₦': 'current_price(₦)', 'Old Price ₦': 'old_price(₦)', '% discount':'(%)discount','Rating' :'rating','Review Count': 'review_count','Shipped?': 'shipped','Official Store?':'official_store'})

df.head()
```

```python
#remove the coma's from the data
def remove_column_comma(columns):
  columns = columns.str.replace(',','',regex= True).astype(float)

  return columns
```

```python
df['old_price(₦)'] = remove_column_comma(df['old_price(₦)'])
```

```python
df['current_price(₦)'] = remove_column_comma(df['current_price(₦)'])
```

```python
#remove the %

def remove_percentage(columns):
  columns = columns.str.replace('%','',regex = True).astype(float)

  return columns
```

```python
df['(%)discount'] = remove_percentage(df['(%)discount'])
```
